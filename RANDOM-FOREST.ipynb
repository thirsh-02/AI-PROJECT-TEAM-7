```python
dataset=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Week4.0-Machine Learning-Classification/Assignment/1. AI in Civil/Concrete_Dataset.csv")
```

```python
dataset
```

```python
independent=dataset[['cement', 'slag', 'flyash','water', 'superplasticizer','coarseaggregate','fineaggregate','age']]
```

```python
independent
```

```python
dependent=dataset[["csMPa"]]
```

```python
dependent
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30,random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(criterion='poisson',max_features='log2',n_estimators = 5000, random_state = 0)
regressor.fit(X_train, y_train)
```

```python
y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```

```python
r_score
```
